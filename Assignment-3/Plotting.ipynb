```python
import matplotlib.pyplot as plt
import torch as th
import torchvision as thv
from torch import nn
from torch.utils.data import DataLoader
from torchvision import datasets, transforms
from torchvision.models import vgg16
```

```python
def plot(source1, source2, title1, title2):

  train = th.load(source1)

  val = th.load(source2)


  fig = plt.figure()
  fig.set_figheight(7)
  fig.set_figwidth(15)
  ax1 = fig.add_subplot(1, 2, 1)
  ax2 = fig.add_subplot(1, 2, 2)
  ax1.set_xticks(range(1, len(train)), minor=False)
  ax2.set_xticks(range(1, len(val)), minor=False)
  ax1.plot(train)
  ax2.plot(val)
  ax1.set_xlabel('Epochs')
  ax1.set_ylabel('Loss')
  ax1.set_title(title1)

  ax2.set_xlabel('Epochs')
  ax2.set_ylabel('loss')
  ax2.set_title(title2)


  plt.show()
```

```python
plot('/content/mean_losses.pt', '/content/val_losses.pt', 'Training Losses', 'Validation Losses')
plot('/content/mean_accuracies.pt', '/content/val_accuracies.pt', 'Training Accuracies', 'Validation Accuracies')
```
